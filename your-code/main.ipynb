# Getting the Audio features of a song

## Building a Data frame of audio features

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import config

# Initialize Spotipy
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(
    client_id=config.SPOTIPY_CLIENT_ID,
    client_secret=config.SPOTIPY_CLIENT_SECRET
))

# Search for Linkin Park tracks
results = sp.search(q="Linkin Park", type="track", limit=50)  # limit is optional

# Extract the list of tracks
tracks = results['tracks']['items']

# Create a list of song names
song_list = [track['name'] for track in tracks]

# Optional: also get artists for each track
artist_list = [track['artists'][0]['name'] for track in tracks]

# Combine into a DataFrame
import pandas as pd
df = pd.DataFrame({
    "Song": song_list,
    "Artist": artist_list
})

print(df)
```

## Searching a playlist

```python
playlist_results = sp.search(q="Linkin Park Hits", type="playlist", limit=5)

playlists = [p for p in playlist_results['playlists']['items'] if p]  # skip None


# Pick the first playlist
playlist = playlists[0]
playlist_id = playlist['id']
playlist_name = playlist['name']

print(f"Playlist: {playlist_name}, ID: {playlist_id}")
```

## Extracting a song from playlist

```python
# Get all tracks from the playlist (handling paging if >100 tracks)
tracks = []
results = sp.playlist_tracks(playlist_id)
tracks.extend(results['items'])

while results['next']:
    results = sp.next(results)
    tracks.extend(results['items'])

# Extract track names and artists
track_list = []
for item in tracks:
    track = item['track']
    track_list.append({
        "Song": track['name'],
        "Artist": track['artists'][0]['name'],
        "URL": track['external_urls']['spotify']
    })

# Convert to DataFrame
df = pd.DataFrame(track_list)
print(df)
```

## Extracting the songs of a playlist

Pagination using "next".
When you collect songs from a playlist using sp.playlist_tracks, you're limited by the limit parameter, which has a maximum (and default) value of 100. When the playlist has more than 100 songs, you have to collect them by navigating through the "pages" of the results.

The parameter offset allows you to retrieve resuls starting at a certain position: if you start at position 101, you'd get the next "page" of results. An offset of 201 would give you the third page, and so on.

The function sp.next() does the same, but in a simpler way: it can be used on the results from any request to directly retrieve the results for the next page.

We can check whether there's a next page or not by accessing the key next on the results from any request.

```python
# Initialize Spotipy
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(
    client_id=config.SPOTIPY_CLIENT_ID,
    client_secret=config.SPOTIPY_CLIENT_SECRET
))

# Search for playlists
playlist_results = sp.search(q="Linkin Park Hits", type="playlist", limit=5)
playlists = [p for p in playlist_results['playlists']['items'] if p]

# Pick the first valid playlist
playlist = playlists[2]
playlist_id = playlist['id']

# Get tracks with paging using sp.next()
results = sp.playlist_tracks(playlist_id)
tracks = results['items']

while results['next']:
    results = sp.next(results)  # automatically fetches the next page
    tracks.extend(results['items'])

# Extract info
track_list = [{"Song": t['track']['name'], "Artist": t['track']['artists'][0]['name']} for t in tracks]
df = pd.DataFrame(track_list)
print(df)
```

### Chopping a big playlist into chunks

```python
# Initialize Spotipy
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(
    client_id=config.SPOTIPY_CLIENT_ID,
    client_secret=config.SPOTIPY_CLIENT_SECRET
))

# Search for playlists
playlist_results = sp.search(q="Linkin Park Hits", type="playlist", limit=5)
playlists = [p for p in playlist_results['playlists']['items'] if p]

# Pick the first valid playlist
playlist = playlists[2]
playlist_id = playlist['id']

# Get tracks with paging using sp.next()
results = sp.playlist_tracks(playlist_id)
tracks = results['items']

while results['next']:
    results = sp.next(results)  # automatically fetches the next page
    tracks.extend(results['items'])

# Extract info
track_list = [{"Song": t['track']['name'], "Artist": t['track']['artists'][0]['name']} for t in tracks]

chunk_1 = []
chunk_2 = []
counter = 0
for song in track_list:
    counter+=1
    if counter <= 50:
        chunk_1.append(song)
    else:
        chunk_2.append(song)
    print(song)


df_1 = pd.DataFrame(chunk_1)
df_2 = pd.DataFrame(chunk_2)
print(df_2)
```
